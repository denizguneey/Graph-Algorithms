```python

import jovian

#Visualize Graphs
import networkx as nx 
import matplotlib.pyplot as plt 

class GraphVisualization: 

    def __init__(self): 

        self.visual = [] 

    def addEdge(self, a, b): 
        temp = [a, b] 
        self.visual.append(temp) 

    def visualize(self): 
        G = nx.Graph() 
        G.add_edges_from(self.visual) 
        nx.draw_networkx(G) 
        plt.show() 

G = GraphVisualization() 
G.addEdge(0, 1) 
G.addEdge(0, 4) 
G.addEdge(1, 2) 
G.addEdge(1, 3) 
G.addEdge(1, 4) 
G.addEdge(2, 3) 
G.addEdge(3, 4) 
G.visualize() 
```
